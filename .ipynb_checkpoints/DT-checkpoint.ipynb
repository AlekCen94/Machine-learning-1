```python
#Import sekcija
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
#Ucitavanje podataka
data = pd.read_csv("heart.csv");
```

```python
data
```

```python
X_data = data.iloc[:,:13].values;
Y_data = data.iloc[:,13].values;
Y_data
```

```python
X_data_train, X_data_test, Y_data_train, Y_data_test = train_test_split(X_data, Y_data,test_size = 1/2)
```

```python
clssifier = DecisionTreeClassifier(max_depth=3)
clssifier.fit(X_data_train, Y_data_train)
Y_pred = clssifier.predict(X_data_test)
accuracy_score(Y_data_test,Y_pred)
```
