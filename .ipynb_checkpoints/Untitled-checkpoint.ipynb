```python
#Import sekcija
import numpy as np
import pandas as pd
import seaborn as sbn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

# Opis skupa podataka

Heart Disease skup podataka kojim se opisuje stanje pacijenata koji boluju od srane bolesti tj. pokazuje uz pomo parametara krvne slike, godina starosti i pola da li je pacijent srani bolesnik ili ne.


```python
#Heart Disease dataset > Funkcija za itanje skupa podataka
def getHeartDiseaseDataset():
        return pd.read_csv("heart.csv");
```

```python
#itanje Heart Disease skupa podataka i prikaz
data = getHeartDiseaseDataset()
data
```

U tabeli iznad moemo videti 13 parametara pomou kojih e model klasifikovati pacijente da li spadaju u srane bolesnike ili ne. Poslednji 14. parametar #target je zapravo labela koja pokazuje da li je pacijent srani bolesnik. 1-Pacijent je srani bolesnik, 0-Pacijent nije srani bolesnik.
Prvih 13 parametara skupa podataka su:

1. age Godine pacijenta.

2. sex Pol pacijenta: 1-muko 0-ensko

3. cp Chest-pain type: Tip bola u grudima
    1-tipina angina
    2-atipina angina
    3-anginalni bol
    4-asiptomatini
    
4. trestbps Vrednost krvnog pritiska u mirovanju pojedinog pacijenta.

5. chol Predstavlja holesterol.

6. fbs Uporeuje

```python
#Priprema podataka > Funkcija koja odvaja labelu target labelu od parametara-obeleija koja su nam od interesa
def prepareData(data):
        X_data = data.iloc[:,:13].values; #Podaci
        Y_data = data.iloc[:,13].values; #Labele
        return X_data, Y_data
```

```python
#Podela skupa podataka na podatke za trenig i testiranje modela.
def splitData(split_size, X_data, Y_data, random_state):
        X_data_train, X_data_test, Y_data_train, Y_data_test = train_test_split(X_data, Y_data,test_size = split_size,random_state = random_state)
        return X_data_train, X_data_test, Y_data_train, Y_data_test
```

```python
#Skaliranje vrednosti obeleija na opseg [-1, 1]
def scaleDataFeatures(X_train, X_test):
        sc = StandardScaler()
        X_tr = sc.fit_transform(X_train)
        X_te = sc.fit_transform(X_test)
        return X_tr, X_te
```

```python
#Klasifikator K-najbliih suseda > Funkcija koja vraa istreniran KN klasifikator
def KNclassifier(X_tr, Y_tr, nn, metrics):
        classifier =  KNeighborsClassifier(n_neighbors = nn, metric = metrics)
        classifier.fit(X_tr, Y_tr)
        return classifier
```

```python
#Predvianje za KN klasifikator > Funkcija koja vri predikciju KN klasifikatora 
def KNpredict(X_te):
        Y_pred = classifier.predict(X_te)
        return Y_pred
```

# Klasifikacja metodom K najbliih suseda
Metoda K najbliih suseda se zasniva na jednom od osnovnih algoritama prepoznavanja uzoraka iji su koraci sledei:
1. Za novi uzorak na osnovu obeleija i odabrane metrike proraunava slinost(daljinu) izmeu novog uzorka i elemenata u koji su rasporeeni.
2. Distanca izmeu novog uzorka i svakog od elemenata stavlja u polje i sortira.
3. iz sortirane liste se uzima prvih k elemenata.
4. Potom se od k elemenata prebrojava maksimalan broj elemenata koji pripadaju istoj grupi
5. Na kraju se novi uzorak dodeljuje grupi koja od prvih k elemenata u polju ima najvie pripadnika. 

# Klasifikacija za K = 5

```python
#Klasifikacija
data = getHeartDiseaseDataset() #itanje podataka
x_data, y_data = prepareData(data) #Priprema podataka
x_train, x_test, y_train, y_test = splitData(0.2, x_data, y_data, 1) #Podela skupa podataka na skup za trening i skup za test.
x_strein_data, x_stest_data = scaleDataFeatures(x_train, x_test) #Skaliranje podataka radi breg izraunavanja udaljenosti.
Kclassifier = KNclassifier(x_strein_data, y_train, 5, 'euclidean') #KNklasifikator
y_predict = Kclassifier.predict(x_stest_data) #Predikcija
accurancy = metrics.accuracy_score(y_test, y_predict) #Proraun tansoti
print('Tanost: ', accurancy)
```

Kao to vidimo iz predhodnih primera, tanost se moe poveati promenom broja uzetih suseda i promenom metrike. Kako bi smo saznali koji je broj suseda najoptimalni za na model koristemo Cross-validation metodu koja e u obzir uzimati broj suseda.

# Cross-validation metoda
je metoda kojom se vri treniranje modela u ovom sluaju KNN klasifikatora za svaku moguu vrednost, u ovom sluaju broja suseda k i proraunava tanost modela. Najoptimalnija vrednost za broj suseda k e biti vrednost za koju model daje najveu tanost.

```python
#Metoda za Cross validaciju.
def KNeighborsCrossValidation(neighbors):
    cross_scores = []
    data = getHeartDiseaseDataset() #itanje podataka
    x_data, y_data = prepareData(data) #Priprema podataka
    x_train, x_test, y_train, y_test = splitData(0.2, x_data, y_data, 1) #Podela skupa podataka na skup za trening i skup za test.
    x_strein_data, x_stest_data = scaleDataFeatures(x_train, x_test) #Skaliranje podataka radi breg izraunavanja udaljenosti.
    
    for k in neighbors:
        
        classifier = KNclassifier( x_strein_data, y_train, k, 'euclidean')
        scores = cross_val_score(classifier,  x_strein_data, y_train)
        cross_scores.append(scores.mean())
    
    return cross_scores

neighbors = list(range(1, 100, 2))
cs = KNeighborsCrossValidation(neighbors)
cs
```
